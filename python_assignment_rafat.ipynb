Question 4: Conditional Statements - (a)Age Checker / (b)Even / Odd (One line) /(c) Grade Calculator / (d) Login System

```python
age = int(input("Enter your age: "))

if age < 0:
    print("Invalid age")
elif age >= 18:
    print("You are an adult")
elif age >= 13:
    print("You are a teenager")
else:
    print("You are a child")     #Age Checker
```

```python
num = int(input("Enter a number: "))
print("Even" if num % 2 == 0 else "Odd")     #Even/Odd
```

```python
mark = int(input("Enter your mark: "))

if mark < 0 or mark > 100:
    print("Invalid mark")
elif mark >= 90:
    print("A+")
elif mark >= 80:
    print("A")
elif mark >= 70:
    print("B")
elif mark >= 60:
    print("C")
elif mark >= 50:
    print("D")
else:
    print("F - Failed")     #grade calculator
```

Question 5: ATM Simulation

```python
correct_username = "admin"
correct_password = "1234"

attempts = 3

while attempts > 0:
    username = input("Enter username: ")
    password = input("Enter password: ")

    if username == correct_username and password == correct_password:
        print("Login successful! Welcome,", username)
        break
    elif username == correct_username:
        print("Incorrect password.")
    else:
        print("User not found.")

    attempts -= 1
    print("Attempts remaining:", attempts)

if attempts == 0:
    print("Account locked")      #Login system
```

Question 6: Basic Loops / (a) For Loop / (b) While Loop

```python
# 1 to 20
for i in range(1, 21):
    print(i)

# 7 multiplication table
for i in range(1, 11):
    print("7 x", i, "=", 7 * i)

# Sum 1 to 100
total = 0
for i in range(1, 101):
    total += i
print("Sum:", total)

# 1 to 50 step 5
for i in range(1, 51, 5):
    print(i)

fruits = ['apple', 'banana', 'cherry', 'mango', 'grape']      #For Loop
for index, fruit in enumerate(fruits, start=1):
    print(index, fruit)
```

```python
# 10 to 1
i = 10
while i >= 1:
    print(i)
    i -= 1

# Countdown
i = 10
while i >= 0:
    if i == 0:
        print("Blast off!")
    else:
        print(i)
    i -= 1

# First 10 multiples of 6
count = 1
while count <= 10:
    print(6 * count)
    count += 1         #While loop
```
